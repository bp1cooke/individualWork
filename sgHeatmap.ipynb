```r
#install.packages("gplots")
library(gplots)
library(RColorBrewer)

myData <- read.csv("math111L.csv")  # read csv file 
#summary(myData)
myData <- myData[,5:6]
#summary(myData$Day)
#summary(myData$Time)
times <- union(myData$Time, myData$Time)
stimes <- sort(times)
#stimes
M <- subset(myData, Day %in% 'Monday')
T <- subset(myData, Day %in% 'Tuesday')
W <- subset(myData, Day %in% 'Wednesday')
R <- subset(myData, Day %in% 'Thursday')
F <- subset(myData, Day %in% 'Friday')

myM <- table(M$Time)
myT <- table(T$Time)
myW <- table(W$Time)
myR <- table(R$Time)
myF <- table(F$Time)
myTemp <- data.frame(myM,myT,myW,myR,myF)
#myTemp
myHeat <- data.frame(myTemp$Var1,myTemp$Freq,myTemp$Freq.1,myTemp$Freq.2,myTemp$Freq.3,myTemp$Freq.4)
#myHeat
names(myHeat)[1] <- "Time"
names(myHeat)[2] <- "Monday"
names(myHeat)[3] <- "Tuesday"
names(myHeat)[4] <- "Wednesday"
names(myHeat)[5] <- "Thursday"
names(myHeat)[6] <- "Friday"
myMat <- as.matrix(myHeat[,2:6])
rownames(myMat) <- myHeat$Time
myCols <- brewer.pal(9,"Blues")

#png(file="heatmap111L.png",width=600,height=600)

heatmap.2(myMat, Rowv = FALSE, Colv = FALSE, dendrogram = 'none', col= myCols)

#dev.off()
```

```r
library(tidyverse)

conflict <- read_csv("math112.csv", col_types = cols(Time=col_time(format="%H:%M:%S")))

png(file="heat112j.png",width=600,height=600)
ggplot(conflict) + 
  geom_bin2d(aes(factor(Day, levels=c("Monday","Tuesday","Wednesday","Thursday","Friday")), 
                 Time), binwidth=300) + 
  scale_y_reverse(breaks=seq(21600, 75600, by=7200), labels=c("6am","8am","10am","12pm","2pm","4pm","6pm","8pm")) + 
  scale_fill_distiller(palette="Blues", direction=1, name="Conflicts (112)") + 
  xlab("Day") +
  theme_bw()

dev.off()
```
